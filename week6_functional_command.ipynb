*Obtaining occurrence data/what my csv contains: Occurrences are records of the presence of a taxon at a particular time and place. The temporal duration of a taxon is known from the oldest and youngest occurrence, and the geographic range is known from the geographic spread of the occurrences.  A typical occurrence record has fields for the taxonomy, the site name, the geological age, and the location.

*Next I will further extract some of the data in this csv to plug into Pyrate (program to estimate speciation, extinction, and preservation rates from fossil occurrence data) in the form of a list. 

*Pyrate will require me to have extracted: 1)species, 2)minimum of data at a given time, and 3)maximum of data at a given time. 

*Therefore, this code for week 6 will allow me to set up these three elements in a way so that I can imput them into pyrate and hopefully acquire my desired speciation rates!

````python
import csv

with open("/home/eeb177-student/Desktop/eeb-177/final-project/formatted_data_frog.csv", newline='') as infh:
    reader = csv.reader(infh)
    counter = 0
    for row in reader:
        #Since my ultimate goal is to set up pyrate to help me get speciation rates, I would like to only focus on species names and therefore will add this to my code:
        if row[6] == "species":
            print(str(counter) + ")" + row[5] + " / " + row[14] + " / " + row[15])
            counter += 1
````
